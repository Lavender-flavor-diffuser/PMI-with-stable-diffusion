```python
# Import necessary libraries
import torch
import matplotlib.pyplot as plt

label = 4

index = 15

# Define the paths to the saved images
image_path = './sampled_images2_label_4.pt'

# Load the images from the saved file with CPU mapping
sampled_images = torch.load(image_path, map_location=torch.device('cpu'))

# Function to display the images
def show_images(images, num_images=100, cols=10):
    rows = num_images // cols
    fig, axes = plt.subplots(rows, cols, figsize=(15, 15))
    for i in range(rows):
        for j in range(cols):
            idx = i * cols + j
            if idx < 1:
                img = images[idx].permute(1, 2, 0).numpy()  # Permute dimensions to (H, W, C) for plotting
                axes[i, j].imshow((img * 0.5 + 0.5))  # Rescale images from [-1, 1] to [0, 1]
                axes[i, j].axis('off')
    plt.show()

# Function to display the images
def show_images2(sampled_images, num_images=100, cols=10):
    rows = num_images // cols
    fig, axes = plt.subplots(rows, cols, figsize=(15, 15))
    for i in range(rows):
        for j in range(cols):
            idx = i * cols + j
            if idx < 100:
                img = sampled_images[idx][index].permute(1, 2, 0).numpy()  # Permute dimensions to (H, W, C) for plotting
                axes[i, j].imshow((img * 0.5 + 0.5))  # Rescale images from [-1, 1] to [0, 1]
                axes[i, j].axis('off')
    plt.show()

show_images2(sampled_images)
show_images2(sampled_images[100:])
show_images2(sampled_images[200:])
show_images2(sampled_images[300:])
show_images2(sampled_images[400:])

# Display the sampled images
# for images in sampled_images:
#     show_images(images)
```
