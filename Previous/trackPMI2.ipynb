```python
# import torch
import numpy as np
import matplotlib.pyplot as plt
import torch

label = 4

index = 15

# 저장된 파일 경로
file_path = '../outputs/PMI_query_Ann Graham Lotz_iter_num_0_sample_num_0.pt'

# 텐서 로드
est_list = torch.load(file_path)


time_steps = list(range(1000, 50 - 1, -50))  # 시간 단계 리스트
plt.plot(time_steps, np.log(est_list), label=f'Index {index}')

plt.xlabel('Time Steps')
plt.ylabel('PMI')
plt.title(f'PMI along Generation for Label ' + str(label))
plt.legend()
plt.show()
```
