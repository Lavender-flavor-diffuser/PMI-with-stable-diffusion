```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# 저장된 파일 경로
file_path = './PMI_label_1.pt'

# 텐서 로드
est_list_list = torch.load(file_path).numpy()

# 각 인덱스에 대해 그래프를 그립니다.
for index, est_list in enumerate(est_list_list):
    time_steps = list(range(3750, 250 - 1, -250)) + [0]  # 시간 단계 리스트
    plt.plot(time_steps, est_list, label=f'Index {index}')

plt.xlabel('Time Steps')
plt.ylabel('PMI')
plt.title(f'PMI along Generation for Label 1')
plt.legend()
plt.show()
```
